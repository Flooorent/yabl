```python
df = pd.DataFrame(data={
            'read': [1, 1, 1, 0, 1, 0, 0],
            'tags': ['roman', 'roman, classique', 'roman', 'roman', 'psycho', 'classique', 'bd'],
        })
df
```

```python
df['x'] = df['tags'].apply(lambda tags: [tag.strip() for tag in tags.split(',')])
df
```

```python
df.explode('x')[['read', 'x']]
```

```python
df.explode('x')[['read', 'x']].groupby('x')['read'].value_counts()
```

```python
y = df.explode('x')[['read', 'x']].groupby('x').agg({'read': ['sum', 'count']})
y
```

```python
y.to_dict()
```

```python
z = y['read'].reset_index()
z
```

```python
z['pct'] = z['sum'] / z['count']
z
```

```python
zz = z[['x', 'count', 'pct']]
zz
```

```python
zz.set_index('x', inplace=True)
```

```python
k = zz.to_dict(orient='index')
k
```

```python
sorted(k, key=lambda keyy: k[keyy]['count'], reverse=True)
```
