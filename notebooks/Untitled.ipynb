```python
import pandas as pd
```

```python
books = pd.read_csv('../data/books_list_2020_02_02_02.csv')
books
```

```python
books['read'].value_counts() / len(books)
```

```python
non_real_books = ['manga', 'bd', 'comics']
real_books = books[~books['tags'].isin(non_real_books)]
real_books
```

```python
real_books['read'].value_counts() / len(real_books)
```

```python
nb_reads_per_tag = books.groupby(['tags', 'read']).count()
nb_reads_per_tag
```
